```
cols = ["TARGET_deathRate", "avgAnnCount", "avgDeathsPerYear", "incidenceRate", "medIncome", "popEst2015", "povertyPercent", "studyPerCap",
        "MedianAge", "AvgHouseholdSize", "BirthRate"]
```

There are no missing values and all the variables are of appropriate datatype.

However, there may be situations where there may be missing values present in the dataset. In that case, they need to be imputed with the mean or the median value, whichever is suitable. The datatypes also need to be converted to the appropriate ones.

```
data.describe()
```

It is to be noted that there are columns whose range (minimum, maximum) vary a lot. For eg., TARGET_deathRate lies between [0, 1] and BirthRate lies between [0, 21.32].

So, we need to scale all the variables to the same range.



```
data_ = (data - data.mean())/data.std()
data_.head()
```

```
data_.describe()
```

Now, all the columns have a mean value ~0 and std deviation of 1. This means that all the values lie between [-1, 1].

The correlation coefficients between the different variables are observed.

```
data_.corr()
```

Observations : 



*   popEst2015 and avgDeathperyear strong positive
*   average annual count and popEst2015 has a high corelation positive
*   povertyPercent and incidenceRate	-> positive
*   medianAge and birthRate is highly negative
*   TARGET_deathRate and MedianAge is very low positive
*   medIncome and MedianAge highly negative
